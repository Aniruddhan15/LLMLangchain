Text To Text

```python
docs = SimpleDirectoryReader("../Data-forPractise")
```

```python
docs1 = docs.load_data()
```

```python
from llama_index.embeddings.gemini import GeminiEmbedding
# get API key and create embeddings

model_name = "models/embedding-001"

embed_model = GeminiEmbedding(
    model_name=model_name, api_key=Google_Api, title="this is a document"
)
```

```python

from llama_index.llms.gemini import Gemini

llm = Gemini(
    model="models/gemini-1.5-flash",
    # api_key="some key",  # uses GOOGLE_API_KEY env var by default
    api_key=Google_Api
)
```

```python
from llama_index.core import Settings
```

```python
service_context = ServiceContext.from_defaults(llm=llm,embed_model=embed_model,chunk_size=1000,chunk_overlap=200)
```

```python
index = VectorStoreIndex.from_documents(docs1,service_context=service_context)
```

```python
index = VectorStoreIndex.from_documents(docs1,service_context=service_context)
```

```python
#Local data base is created and docs1 embeddings is stored along with index

index.storage_context.persist() 
```

# Chat system -  retain the memory, Query system - is for just retrieval

```python
q_engine = index.as_query_engine()
```

```python
res = q_engine.query("how many months does it take to learn android dev with kotlin")
```
